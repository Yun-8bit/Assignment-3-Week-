```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 불러오기
X_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/boston_house_data.xlsx')
y_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/boston_house_target.xlsx')

# 입력 데이터의 마지막 열, 출력 데이터의 2번째 열 사용
X = X_data.iloc[:, -1].values.reshape(-1, 1)
y = y_data.iloc[:, 1].values.reshape(-1, 1)

# 학습/테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 데이터 정규화
X_mean = X_train.mean()
X_std = X_train.std()
X_train_norm = (X_train - X_mean) / X_std
X_test_norm = (X_test - X_mean) / X_std

# 경사하강법 구현 (다항회귀)
def gradient_descent_poly(X, y, learning_rate=0.0001, iterations=10000):
    m, n = X.shape
    w = np.zeros((n, 1))
    for _ in range(iterations):
        y_pred = X.dot(w)
        dw = (-2/m) * X.T.dot(y - y_pred)
        w -= learning_rate * dw
    return w

degrees = [2, 3, 4]

plt.figure(figsize=(18, 18))

for i, degree in enumerate(degrees, 1):
    # 다항 특성 변환
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train_norm)
    X_test_poly = poly.transform(X_test_norm)

    # 경사하강법으로 학습
    w_gd = gradient_descent_poly(X_train_poly, y_train)
    y_pred_gd = X_test_poly.dot(w_gd)

    # Scikit-learn으로 학습
    lr_model = LinearRegression()
    lr_model.fit(X_train_poly, y_train)
    y_pred_lr = lr_model.predict(X_test_poly)

    # 정확도 평가
    mse_gd = mean_squared_error(y_test, y_pred_gd)
    r2_gd = r2_score(y_test, y_pred_gd)

    mse_lr = mean_squared_error(y_test, y_pred_lr)
    r2_lr = r2_score(y_test, y_pred_lr)

    # 시각화 (GD)
    plt.subplot(3, 2, 2*i-1)
    sorted_idx = X_test_norm[:, 0].argsort()
    plt.scatter(X_test_norm, y_test, color='blue', label='Actual')
    plt.plot(X_test_norm[sorted_idx], y_pred_gd[sorted_idx], color='red', linewidth=2, label='GD Prediction')
    plt.title(f'GD Polynomial (Degree {degree})\nMSE: {mse_gd:.2f}, R²: {r2_gd:.3f}')
    plt.xlabel('Normalized Feature')
    plt.ylabel('Price')
    plt.legend()

    # 시각화 (Scikit-learn)
    plt.subplot(3, 2, 2*i)
    plt.scatter(X_test_norm, y_test, color='blue', label='Actual')
    plt.plot(X_test_norm[sorted_idx], y_pred_lr[sorted_idx], color='green', linewidth=2, label='Sklearn Prediction')
    plt.title(f'Sklearn Polynomial (Degree {degree})\nMSE: {mse_lr:.2f}, R²: {r2_lr:.3f}')
    plt.xlabel('Normalized Feature')
    plt.ylabel('Price')
    plt.legend()

plt.tight_layout()
plt.show()
```
